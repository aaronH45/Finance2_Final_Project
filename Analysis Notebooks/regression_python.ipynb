```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from linearmodels.panel import PanelOLS
from linearmodels.iv import IV2SLS
from pystout import pystout
from statsmodels.iolib.summary2 import summary_col
```

```python
df_grouped = pd.read_pickle("./reg_df.pkl")  
df_grouped
```

```python
df_grouped.describe()
```

```python
exog_vars = df_grouped[['rescaled_similarity_lagged']]
exog_vars = sm.add_constant(exog_vars)
mod = PanelOLS(df_grouped.rescaled_similarity, exog_vars, time_effects=True)
res_total = mod.fit(cov_type='clustered', cluster_entity=True)
res_total
```

```python
pystout([res_total], file='output.txt', digits = 2)
```

```python
exog_vars = df_grouped[['rescaled_similarity']]
exog_vars = sm.add_constant(exog_vars)
mod = PanelOLS(df_grouped.spread_change*100, exog_vars, time_effects=False)
res_total1 = mod.fit(cov_type='clustered', cluster_entity=True)


exog_vars = df_grouped[['rescaled_similarity']]
exog_vars = sm.add_constant(exog_vars)
mod = PanelOLS(df_grouped.spread_change*100, exog_vars, time_effects=True)
res_total2 = mod.fit(cov_type='clustered', cluster_entity=True)

```
